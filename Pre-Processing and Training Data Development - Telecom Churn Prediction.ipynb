# Pre-Processing and Training Data Development : Telecom Churn Prediction

```python
#load python packages
import os
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

%matplotlib inline
```

```python
data = Path('data/cleaned_data.csv')
# load the data file into a Data Frame
df = pd.read_csv(data)
# inspect the dataframe
df.head()
```

```python
# copy the dataframe into other
df_cleaned = df.copy()
# mapping the data points in SeniorCitizen 
df_cleaned.SeniorCitizen = df.SeniorCitizen.apply(lambda x: 'Yes' if x==1 else 'No')
```

```python
df_cleaned.dtypes
```

```python
dfo=df_cleaned.drop(['customerID','Churn'], axis=1).select_dtypes(include='object')
dfx = pd.concat([df_cleaned.drop(dfo, axis=1), pd.get_dummies(dfo)], axis=1)
```

```python
dfx.head()
```

```python
# first we import the preprocessing package from the sklearn library
from sklearn import preprocessing

# Declare an explanatory variable, called X,and assign it the result of dropping 'customerID' and 'Churn' from the df
X = dfx.drop(['customerID','Churn'], axis=1)

# Declare a response variable, called y, and assign it the Churn column of the df 
y = dfx.Churn 

# Here we use the StandardScaler() method of the preprocessing package, and then call the fit() method with parameter X 
scaler = preprocessing.StandardScaler().fit(X)

# Declare a variable called X_scaled, and assign it the result of calling the transform() method with parameter X 
X_scaled=scaler.transform(X) 
```

```python
corr = X.corr().abs()
corr.round(2).style.background_gradient(cmap='coolwarm')
```

```python
# Select upper triangle of correlation matrix
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
```

```python
to_drop
```

```python
df_processed = pd.concat([X.drop(to_drop, axis=1), y], axis=1)
df_processed.head()
```
